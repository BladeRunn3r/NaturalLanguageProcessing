```python
import nltk
nltk.download('punkt')
nltk.download('universal_tagset')
sentences = ('Add two kitkats for 2 dollars and one mars for 1.5 dollars',
            'Add twwo bunss for 2 dolars and two cqndies for 1.5 dollars')
for sntc in sentences:
    print(nltk.pos_tag(nltk.word_tokenize(sntc), tagset='universal'), '\n')
```

```python
import nltk
nltk.download('punkt')
nltk.download('wordnet')
#lemmatizer - real db
from nltk.stem import WordNetLemmatizer
wnl = WordNetLemmatizer()
sentences = ('Add two KitKats for 2 dollars and one Mars for 1.5 dollars',
            'Add two buns for 2 dollars and two candies for 1.5 dollars')
for sntc in sentences:
    tokens = nltk.word_tokenize(sntc)
    for t in tokens:
        print(wnl.lemmatize(t))
    print('\n')
# doesnt work well with brands, ok without
```

```python
#pyspellchecker
import nltk
from spellchecker import SpellChecker
spell = SpellChecker()
sntc = 'Ad twwo bunss for 2 dolars and two cqndies for 1.5 dollars'
misspelled = spell.unknown(nltk.word_tokenize(sntc))
for word in misspelled:
    # Get the one `most likely` answer
    print(word,'->',spell.correction(word))
    # Get a list of `likely` options
    print(spell.candidates(word))
    print('\n')
# works ok for simplest mistakes
```

```python
import jamspell
corrector = jamspell.TSpellCorrector()
corrector.LoadLangModel('en.bin')
#uses already trained network (60mb binary)

sntc = 'Ad twwo bunss for 2 dolars and two cqndies for 1.5 dollars'
corrector.FixFragment(sntc)
```
